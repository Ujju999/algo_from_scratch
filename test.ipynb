```python
import numpy as np
```

```python
from linear_regression import LinearRegression
```

```python
np.random.seed(12)
X = np.random.rand(1000,2) * 10
true_w = np.array([6.71,-0.91])
y = X @ true_w + 7 + np.random.randn(1000) * 0.5
model = LinearRegression(learning_rate=0.01,tolerance=1e-6,n_steps=8000)
model.fit(X,y)
print("Complete")
```

```python
print(f"Learned Weights: {model.w}")
```

```python
model.plot_loss()
```

```python
np.random.seed(12)

X_class0 = np.random.randn(50,2) + np.array([2,2])
X_class1 = np.random.rand(50,2) + np.array([-2,-2])

X = np.vstack([X_class0,X_class1])
y = np.hstack([np.zeros(50), np.ones(50)])

indices = np.random.permutation(100)
X = X[indices]
y = y[indices]
```

```python
from logistic_regression import LogisticRegression
model = LogisticRegression(tolerance=1e-6,learning_rate=0.01,n_steps=10000)
model.fit(X,y)
```

```python
model.b
```

```python
model.score(X,y)
```

```python
model.loss[-1]
```

```python
model.plot_loss()
```

```python
test_points = np.array([[2,2],[-2,-2],[0,0]])
prob = model.predict_proba(test_points)
```
